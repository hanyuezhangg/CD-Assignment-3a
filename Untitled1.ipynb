```python
import pandas as pd

df = pd.read_csv('/Users/zhanghanyue/Desktop/dataset/Global_Education.csv', delimiter=",", encoding='ISO-8859-1')
```

```python
df
```

```python
top_10_unemployment = df[['Countries and areas', 'Unemployment_Rate']].nlargest(10, 'Unemployment_Rate')

print(top_10_unemployment)
```

```python
top_20_unemployment = df[['Countries and areas', 'OOSR_Upper_Secondary_Age_Female']].nlargest(20, 'OOSR_Upper_Secondary_Age_Female')

print(top_20_unemployment)
```

```python
selected_columns = ['Countries and areas', 'Gross_Primary_Education_Enrollment', 'Gross_Tertiary_Education_Enrollment']
selected_df = df[selected_columns]
top_10_lowest = selected_df.nlargest(10, ['Gross_Primary_Education_Enrollment', 'Gross_Tertiary_Education_Enrollment'])

print(top_10_lowest)
```

```python
filtered_df = df[df['OOSR_Upper_Secondary_Age_Female'] > 70]
dropout_rate = filtered_df[['Countries and areas', 'OOSR_Upper_Secondary_Age_Female']].set_index('Countries and areas')
Birth_Rate = filtered_df[['Countries and areas', 'Birth_Rate']].set_index('Countries and areas')

ax1 = dropout_rate.plot(kind='line', marker='o', title='Female Primary School Dropout Rate and Fertility Rate')
Birth_Rate.plot(ax=ax1, secondary_y=True, marker='x')

ax1.set_ylabel('OOSR_Upper_Secondary_Age_Female')
ax1.right_ax.set_ylabel('Birth_Rate')
```
